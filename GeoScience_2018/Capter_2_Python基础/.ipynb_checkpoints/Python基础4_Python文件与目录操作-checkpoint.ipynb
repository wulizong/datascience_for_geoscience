### 2.3 os.walk()

>函数声明：walk(top,topdown=True,onerror=None)

* 参数top表示需要遍历的目录树的路径
* 参数topdown的默认值是"True",表示首先返回目录树下的文件，然后在遍历目录树的子目录.Topdown的值为"False"时，则表示先遍历目录树的子目录，返回子目录下的文件，最后返回根目录下的文件
* 参数onerror的默认值是"None",表示忽略文件遍历时产生的错误.如果不为空，则提供一个自定义函数提示错误信息后继续遍历或抛出异常中止遍历
* 该函数返回一个元组，该元组有3个元素，这3个元素分别表示每次遍历的路径名，目录列表和文件列表

```python
filepath='C:/WestCode/Jupyter/GeoData/Day_1/'
i =0
for root, dirs, files in os.walk(filepath):
    if i<1:
        print("Root = ", root, "dirs = ", dirs, "files = ", files)
    i=i+1
```

```python
filepath='../'
for root, dirs, files in os.walk(filepath):
    for filespath in files:
      print(os.path.join(root,filespath))
```

## 3.目录操作

* 当前工作目录

```python
os.getcwd()
```

* 判断是否是目录

```python
filepath='../'
if os.path.exists(filepath):
    if os.path.isdir(filepath):
        if not os.path.isabs(filepath):  
            print('绝对路径:',os.path.abspath(filepath))
        else:
            print('不是绝对路径')
    else:
        print('该文件不是路径')
else:
    print('该目录不存在')
```

* 路径分割

```python
filename='data/weather_data.txt'
fullname=os.path.abspath(filename)
print('绝对路径：',fullname)
print('路径分割:',os.path.split(fullname))
print('所在目录：',os.path.dirname(fullname))
print('所在硬盘：',os.path.splitdrive(fullname)[0])
```

* 路径合并

```python
dir_name='c:/temp'
basename='test.shp'
filename=os.path.join(dir_name,basename)
filename
```

## 5.综合应用
遍历一特定数据目录下，获得以下文件信息：
* 所在目录
* 文件名
* 文件大小(以Mb为单位)
* 文件格式格式


```python
import os
file_path='../'
for root,sub_dir,files in os.walk(file_path):
    for filename in files:
        full_path=os.path.abspath(root)
        full_name=os.path.abspath(os.path.join(root,filename))
        print('所在文件目录:',full_path)
        print('文件名:',full_name)
        print('文件大小:',round(os.path.getsize(full_name)/1024.,1),'Mb')
```
