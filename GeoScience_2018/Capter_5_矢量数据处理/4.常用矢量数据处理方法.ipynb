## 1.地图投影转换

### 1.1地图投影基本概念
#### 1）EPSG
* European Petroleum Survey Group (EPSG)， http://www.epsg.org/，它成立于1986年，并在2005年重组为OGP(Internation Association of Oil & Gas Producers)，它负责维护并发布坐标参照系统的数据集参数，以及坐标转换描述，该数据集被广泛接受并使用。
目前已有的椭球体，投影坐标系等不同组合都对应着不同的ID号，这个号在EPSG中被称为EPSG code，它代表特定的椭球体、单位、地理坐标系或投影坐标系等信息。
* 在线查询：http://spatialreference.org/ref/epsg/

#### 2）wkt格式

WKT(Well-known text)是一种文本标记语言，用于表示矢量几何对象、空间参照系统及空间参照系统之间的转换。它的二进制表示方式，亦即WKB(well-known binary)则胜于在传输和在数据库中存储相同的信息。该格式由开放地理空间联盟(OGC)制定。


#### proj4格式

Proj4是一个开源的GIS地图投影库，它专注于地图投影的表达，以及转换。采用一种非常简单明了的投影表达－－PROJ4，比其它的投影定义简单，但很明显。

#### 问题1：哪些气象站位于云南（点数据切割）

```python
#encoding=utf8
import geopandas as gpd
import pandas as pd
import numpy as np
%matplotlib inline
```

* Step1:浏览行政区划数据

```python
shapefile_admin='data/china_boundary_province.shp'
df_admin=gpd.read_file(shapefile_admin,encoding='gbk')
df_admin.head()
```

* 选择出云南的行政边界

```python
bound_yn=df_admin[df_admin['NAME']==u'云南省']
ax=bound_yn.plot(figsize=(8,8))
```

* 浏览全国气象站数据

```python
station='data/weather_station.shp'
df_station=gpd.read_file(station,encoding='utf8')
df_station.head()
```

* 绘制出全国气象站分布图

```python
ax=df_admin.plot(figsize=(8,8),facecolor=None,linewidth=1)
df_station.plot(ax=ax,color='red')
```

* 求出位于云南的气象站

```python
from geopandas.tools import sjoin
station_yn = sjoin(df_station, bound_yn, how="inner",op='within')
station_yn.head()
```

* 绘制出云南气象站分布图

```python
ax1 = bound_yn.plot(figsize=(8,8))
ax2 = station_yn.plot(ax=ax1)
```

* 保存输出云南气象站分布图


```python
station_yn.columns
```

```python
station_yn_new = station_yn[[ 'ID', 'name','latitude', 'longitude','elevation','CLASS', 
                             'date_start', 'DESCRIPT','geometry']]
station_yn_new.head()
```

```python
# 修改输出字段名
station_yn_new.columns=[ 'id', 'name','latitude', 'longitude','elevation','class', 
                             'date_start','descript', 'geometry']
station_yn_new.head()
```

```python
# 修改字段类型
```

```python
station_yn_new['id']=station_yn_new['id'].astype(np.int64)
station_yn_new['elevation']=station_yn_new['elevation']/10.0
station_yn_new.head()
```

```python
station_yn_new.geometry.crs=df_station.crs
station_yn_new.crs
```

```python
from collections import OrderedDict

schema={'geometry': 'Point',
        'properties': OrderedDict([(u'id','int:6'),
                                   (u'name', 'str:30'),(u'latitude', 'float:7.2'),
                                   (u'longitude', 'float:7.2'),
                                   (u'elevation', 'float:7.2'),(u'class', 'str:20'),
                                   (u'date_start', 'str:20'),(u'descript', 'str:40')])}
station_yn_new.to_file('c:/temp/station_yunnan.shp',encoding='utf8',schema=schema)
```
