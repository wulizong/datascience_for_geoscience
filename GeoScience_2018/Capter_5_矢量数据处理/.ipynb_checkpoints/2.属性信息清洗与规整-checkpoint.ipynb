## 1.清洗字段信息


### 1.1清洗字段名称

```python
#encoding=utf8
import pandas as pd
import geopandas as gpd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
shapefile='data/countypolygon_1986.shp'
df = gpd.read_file(shapefile,encoding='gbk')
df.head()
```

```python
df.columns=['clas_id', 'cnty_code', 'fullname', 'name_zh', 'name_py', 'shape_area',\
       'shape_leng', 'code99', 'data_year', 'dq_name', 'geometry', 'note0',\
       'note1', 'note2', 'note3', 'prov_zh']
df.head()
```

### 1.2清洗字段类型

* 将cnty_code和code99改为整形

```python
df.dtypes
```

```python
import numpy as np
df['cnty_code']=df['cnty_code'].astype(np.int64)
df['code99']=df['code99'].astype(np.int64)
df.dtypes
```

### 1.3调整字段顺序

```python
df_new=df[['clas_id','cnty_code','fullname','name_zh',
          'name_py','prov_zh','code99','data_year','dq_name',
          'note0','note1','note2','note3','shape_area','shape_leng','geometry']]
df_new.head()
```

## 2.清洗属性表内容
### 2.1添加新的字段

* 从其他的文件中读取信息

```python
csv_file='data/province_name.csv'
df_name=pd.read_csv(csv_file,encoding='utf8')
df_name.head()
```

* 属性表关联

```python
df_merge=df.merge(df_name,how='left',left_on=['prov_zh'],right_on=['name_zh'])
df_merge.head()
```

* 修改字段顺序

```python
df_merge.columns
```

```python
df_new=df_merge[['clas_id','fullname', 'name_zh_x', 'name_py_x', 'prov_zh',
                'name_py_y','cnty_code','code99', 'data_year', 'dq_name',
                'note0', 'note1', 'note2','note3',
                'shape_area', 'shape_leng', 'geometry']]
       
df_new.columns=['clas_id','fullname', 'name_zh', 'name_py', 'prov_zh','prov_py',
                'cnty_code','code99', 'data_year', 'dq_name', 
                'note0', 'note1', 'note2','note3',
                'shape_area', 'shape_leng', 'geometry']
df_new.head()
```

## 3.属性表内容检查
### 3.1通过统计数，检查每个省县级区划数是否合适

```python
grouped= df_new.groupby(['prov_zh','name_zh']).count().reset_index()
county_count=grouped['prov_zh'].value_counts().reset_index()
county_count.columns=['prov_zh','count']
county_count
```

### 3.2通过画图检查

```python
ax=df_new.plot(column='prov_zh',figsize=(12,12))
plt.savefig('c:/temp/china_county_boundary_1986.png')
```

### 3.3保存清洗结果

```python
df_new.to_file('c:/temp/china_county_boundary_1986.shp',encoding='gbk')
```

## 4.按属性输出数据

### 4.1按省名，输出分省县界数据

```python
df_prov=df_new['prov_py'].value_counts()
```

```python
import os
out_path='c:/temp'
for prov in df_prov.index:
    df_temp=df_new[df_new['prov_py']==prov]
    out_file=os.path.join(out_path,'china_county_boundary_1986_'+prov+'.shp')
    df_temp.to_file(out_file,encoding='gbk')
```
