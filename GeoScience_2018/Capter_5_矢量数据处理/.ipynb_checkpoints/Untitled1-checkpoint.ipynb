```python
import geopandas as gp
import pandas as pd
%matplotlib inline
shapefile='data/countypolygon_1986.shp'
df=gp.read_file(shapefile,encoding='gbk')
df.plot(figsize=(10,10))
df.columns
df['CNTY_CODE']=df['CNTY_CODE'].astype('int64')
df['code99']=df['code99'].astype('int64')
df['new2']=df['data_year']+1000
df['CNTY_CODE']=df['CNTY_CODE'].astype('int64'
df['code99']=df['code99'].astype('int64')
col_name=['Class_ID', 'cnty_code', 'fullname', 'Name', 'Name_py', 'Shape_Area',
       'Shape_Leng', 'Code_99', 'Data_year', 'dq_name', 'geometry', 'note0',
       'note1', 'note2', 'note3', 'name_prov', 'new', 'new2']
df.columns=col_name
df_new=df[['Class_ID', 'cnty_code', 'Code_99',
           'Name','fullname',  'Name_py','name_prov',
           'Data_year', 'dq_name',  
           'note0','note1', 'note2', 'note3',
           'Shape_Area','Shape_Leng', 'geometry']]
filename='data/province_name.csv'
df_1 = pd.read_csv(filename)
df_new2=df_new.merge(df_1,left_on=['name_prov'],right_on=['name_zh'])
df_new2.to_file('c:/temp/new_shape1.shp',encoding='gbk')
```
