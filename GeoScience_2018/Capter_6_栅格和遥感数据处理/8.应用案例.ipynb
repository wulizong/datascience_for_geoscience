##  1.从遥感数据的元数据文件中获取信息
* 以Landsat数据为例

```python
#encoding=utf8
import os
import glob
import geopandas as gp
import pandas as pd
from shapely.geometry import Polygon
import fiona
from fiona.crs import from_epsg
%matplotlib inline
```

* 依然需要了解wkt格式

```python
POINT(6 10)
LINESTRING(3 4,10 50,20 25)
POLYGON((1 1,5 1,5 5,1 5,1 1),(2 2,2 3,3 3,3 2,2 2))
```

```python
filepath='data/landsat_met/'
filelist =glob.glob(os.path.join(filepath,'*.met'))

filelist
```

```python
col_name=['key','value']
df_all=[]
geom_poly=[]
file_id=[]
image_size=[]
spacecraft_id=[]
sensor_id=[]
acquisition_date=[]
wrs_path=[]
wrs_row=[]
for filename in filelist:
    basename=os.path.basename(filename).replace('.met','')
    file_id.append(basename)
    df=pd.read_csv(filename,sep='=',encoding='utf8')
    df.columns=['key','value']
    df['key']=df['key'].str.replace('\t','')
    df['key']=df['key'].str.replace(' ','')
    df.set_index(['key'],inplace=True)
    df_T=df.T
    image_size.append(df_T['PRODUCT_FILE_SIZE'].ix[0])
    spacecraft_id.append(df_T['SPACECRAFT_ID'].ix[0])
    sensor_id.append(df_T['SENSOR_ID'].ix[0])
    acquisition_date.append(df_T['ACQUISITION_DATE'].ix[0])
    wrs_path.append(df_T['WRS_PATH'].ix[0])
    wrs_row.append(df_T['WRS_ROW'].ix[0])
    #geom
    UL_x = float(df_T.SCENE_UL_CORNER_LAT.ix[0])
    UL_y = float(df_T.SCENE_UL_CORNER_LON.ix[0])
    UR_x = float(df_T.SCENE_UR_CORNER_LAT.ix[0]) 
    UR_y = float(df_T.SCENE_UR_CORNER_LON.ix[0]) 
    LL_x = float(df_T.SCENE_LL_CORNER_LAT.ix[0]) 
    LL_y = float(df_T.SCENE_LL_CORNER_LON.ix[0]) 
    LR_x = float(df_T.SCENE_LR_CORNER_LAT.ix[0]) 
    LR_y = float(df_T.SCENE_LR_CORNER_LON.ix[0]) 
    poly=Polygon([(UL_x,UL_y),(UR_x,UR_y),(LR_x,LR_y),(LL_x,LL_y),(UL_x,UL_y)])       
    geom_poly.append(poly) 

#GeoSeries
geometry=gp.GeoSeries(geom_poly)
geometry.crs=from_epsg(4326)
df_landsat = gp.GeoDataFrame(file_id,columns=['file_id'])
df_landsat['geometry']=geometry
df_landsat['image_size']=image_size
df_landsat['craft_id']=spacecraft_id
df_landsat['sensor_id']=sensor_id
df_landsat['acqu_date']=acquisition_date
df_landsat['wrs_path']=wrs_path
df_landsat['wrs_row']=wrs_row

df_landsat.to_file('c:/temp/landsat_index_map.shp')
```

```python
df_landsat.plot(figsize=(12,12))
```

## 4. rasterio

```python
import numpy as np
import rasterio
import matplotlib.pylab as plt
from rasterio import plot
%matplotlib inline
```

```python
filename='data/raster/s_temp0cm/hdr.adf'
src = rasterio.open(filename)
```

```python
#help(src)
```

* 元数据

```python
src.meta
```

* 基本信息

```python
print('文件名:',src.name)
print("格式:",src.driver)
print("图像大小:",src.height, src.width)
print("波段数:",src.count)
print('地图投影(proj4):',src.crs_wkt)
print('分辨率',src.res)
print('数据类型',','.join(src.dtypes))
print('Nodata:',src.nodata)
print('空间范围:',src.bounds) 
```

* 基本统计信息

```python
print('统计信息:')
for i in range(src.count):
    band=src.read(i+1,masked=True)
    print("band_%s:'min':%f,'max':%f,'mean':%f"%
          (str(i+1),band.min(),band.max(), band.mean()))
```

```python
band.std()
```

```python
fig=plt.figure(figsize=(7,7))
plot.show(band)
```

* DN 值统计直方图

```python
plot.show_hist(band,bins=20)
```
