### 3.绘制缩略图

```python
import cartopy.crs as ccrs
from mpl_toolkits.axes_grid1 import make_axes_locatable

with rasterio.open(filename) as src:
    im = src.read()
    xmin = src.transform[0]
    xmax = src.transform[0] + src.transform[1]*src.width
    ymin = src.transform[3] + src.transform[5]*src.height
    ymax = src.transform[3]
    im[im == -3.40282346639e+38] = np.nan
    fig=plt.figure(figsize=(20,20))

    my_crs = ccrs.AlbersEqualArea(central_longitude=105, central_latitude=0.0, false_easting=0.0, \
                              false_northing=0.0, standard_parallels=(25.0, 47.0), globe=None)
    ax = plt.axes(projection=my_crs)
    plt.imshow(im[0], origin='upper', extent=[xmin, xmax, ymin, ymax], \
               transform=my_crs, interpolation='nearest',vmin=-3, vmax=24,cmap='gray')
    #divider = make_axes_locatable(ax)
    #cax = divider.append_axes("right", size="5%", pad=0.05)

    #plt.colorbar()
    ax.gridlines(crs=my_crs)
    
    plt.show()
    #fig.canvas.draw()
```

```python
## 为便于制图，将地图投影转换
## 转换为经纬度投影
## 转换后，求地图范围
import numpy as np
import rasterio
from rasterio import Affine as A
from rasterio.warp import calculate_default_transform,reproject, RESAMPLING

dst_crs = 'EPSG:4326'
xmin = 70
xmax = 140
ymin = 5
ymax = 55
filename='C:/Geodata/dataset/raster/s_temp0cm/hdr.adf'
with rasterio.open(filename) as src:
    affine, width, height = calculate_default_transform(
        src.crs, dst_crs, src.width, src.height, *src.bounds)
    kwargs = src.meta.copy()
    kwargs.update({
        'crs': dst_crs,
        'transform': affine,
        'affine': affine,
        'width': width,
        'height': height
    })
    
    band=src.read()[0]
    dest = np.zeros_like(band)
    reproject(
                source=band,
                destination=dest,
                src_transform=src.affine,
                src_crs=src.crs,
                dst_transform=affine,
                dst_crs=dst_crs,
                resampling=RESAMPLING.nearest)

    
    fig=plt.figure(figsize=(20,20))
    ax = plt.axes(projection=ccrs.PlateCarree())
    plt.imshow(dest, origin='upper', extent=[xmin, xmax, ymin, ymax], \
        transform=ccrs.PlateCarree(),interpolation='nearest',vmin=-3, vmax=24,cmap='gray')
    ax.set_xlim(70,140)
    ax.set_ylim(5,55)
    ax.gridlines(crs=ccrs.PlateCarree(),draw_labels=True)    
    plt.show()
```

```python
## 分省制图
import numpy
import fiona
import rasterio
import rasterio.features
from affine import Affine
from shapely.geometry import shape
import geopandas as gpd
df = gpd.read_file('c:/geodata/China_boundary_province.shp',encoding='utf8')
print len(df[df['NAME']==u'甘肃省'])
with fiona.open('c:/geodata/China_boundary_province.shp', 'r') as vector:
    with rasterio.open('C:/Geodata/dataset/raster/s_temp0cm/hdr.adf', 'r') as raster:
        #feature=vector[vector['properties']['NAME']== u'甘肃省']
        for feature in vector:
            if feature['properties']['ENTIID']== '1CHN620000000':
                geometry = shape(feature['geometry'])
                # get pixel coordinates of the geometry's bounding box
                
                ul = raster.index(*geometry.bounds[0:2])
                lr = raster.index(*geometry.bounds[2:4])
                print ul,lr
                #投影

        
    
```

```python
print len(features)
        

for feature in vector:
        # create a shapely geometry
        # this is done for the convenience for the .bounds property only
        

        # read the subset of the data into a numpy array
        window = ((lr[0], ul[0]+1), (ul[1], lr[1]+1))
        data = raster.read_band(1, window=window)

        # create an affine transform for the subset data
        t = raster.affine
        shifted_affine = Affine(t.a, t.b, t.c+ul[1]*t.a, t.d, t.e, t.f+lr[0]*t.e)

        # rasterize the geometry
        mask = rasterio.features.rasterize(
            [(geometry, 0)],
            out_shape=data.shape,
            transform=shifted_affine,
            fill=1,
            all_touched=True,
            dtype=numpy.uint8)

        # create a masked numpy array
        masked_data = numpy.ma.array(data=data, mask=mask.astype(bool))

        # display some statistics
        name = feature['properties']['NAME'].title()
        print('{}\nRange: {:.1f} to {:.1f}'.format(
            name,
            masked_data.min(),
            masked_data.max()))
```
