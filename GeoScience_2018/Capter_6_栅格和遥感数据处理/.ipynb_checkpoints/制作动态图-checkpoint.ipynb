## MODIS Snow

```python
import numpy as np
import rasterio
import matplotlib.pylab as plt
from rasterio import plot
%matplotlib inline
```

```python
filename='data/Modis_Snow/MOD10A1.A2000059.Snow_Cover_Daily_Tile.tif'
src = rasterio.open(filename)
```

```python
src.meta
```

```python
band=src.read(1)
```

```python
band.min(),band.max()
```

```python
plot.show(band,cmap='winter')
```

```python
plot.show_hist(band)
```

```python
band[band>100]=255
band[band<200]=0
```
