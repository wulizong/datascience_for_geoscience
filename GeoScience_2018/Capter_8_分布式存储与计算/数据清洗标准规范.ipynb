<h1 align=center>数据清洗方案</h1>

## 1.背景
 * 数据中心的数据正遭到原来越多的恶意下载，
 * 数据粒度过大，无法有效控制数据下载范围
 * 数据拆分过程中存在大量的重复工作，如拆分数据实体，撰写元数据、撰写数据文档、制作缩略图、上传数据，发布数据等
 * 

## 2.实施步骤

### 2.1 第一步
* 解决解决当前最紧要的数据恶意下载问题
* 开发用户注册认证功能
* 实现第一步数据拆分，将数据粒度变小
* 恢复正常的数据服务

### 2.2 第二步
* 以数据为核心，开展彻底数据清洗 
* 开发新的系统，从元数据为中心的系统向以数据为中心的系统转化
* 利用大数据和云技术，实现在数据共享、分析和可视化

## 3.实施方案

### 3.1 数据摸底调查 
* 已经开展
* 进一步确认，哪些数据是热门各个数据中心的热门数据，
* 优先针对热门数据开展数据拆分





### 3.2 第一阶段数据拆分方案
* 以减小数据粒度，控制数据下载为主要目的
* 以数据清洗侧重于数据格式和数据命名规范的统一
* 同步清洗元数据和数据文档
* 统一撰写相关数据处理工具，减少重复体力劳动
* 充分利用原有系统开展数据拆分工作


#### 3.2.1 观测数据的拆分和清洗原则
* 暂不进行数据格式的统一
* 暂不进行数据内容的清洗
* 数据实体按“观测站和观测要素”进行拆分，或者按“观测站+观测年代”的方式进行拆分
* 每一个观测站作为一个元数据单元（不同数据中心可根据自己的实际情况酌情变更）
* 每一个元数据关联若干数据实体。但数据实体文件，最多有一个在线下载，其他数据需要离线申请
* 结合用户控制系统进行数据下载控制
* 批量形成元数据
* 批量形成数据文档

#### 3.2.2 统计数据的拆分和清洗原则
* 暂不进行数据格式的统一
* 暂不进行数据内容的清洗
* 数据实体的拆解：
 * 按照行政区（如县）+统计要素进行拆分
* 统计数据按如下要素进行拆分
 * 综合
 * 国民经济核算
 * 人口
 * 就业人员和工资
 * 固定资产投资和房地产
 * 对外经济贸易
 * 能源
 * 财政
 * 价格指数
 * 人民生活
 * 城市概况
 * 资源和环境
 * 农业
 * 工业
 * 建筑业
 * 运输和邮电
 * 社会消费品零售总额
 * 批发和零售业
 * 住宿和餐饮业
 * 旅游业
 * 金融业
 * 教育
 * 科技
 * 卫生
 * 社会服务
 * 文化
 * 体育
 * 公共管理、社会保障及其他
* 数据拆分需要手工拆解
* 尽量避免要求重新上传数据
* 尽量避免严重破坏原有数据结构
* 批量撰写元数据
* 批量撰写数据文档

#### 3.2.3 矢量数据拆分和清洗原则

* 400万以上数据，以国家为单位拆分
* 100万-400万数据以省为单位拆分
 * 可以以省为单位进行拆分
 * 也可以以要素或图层为单位进行拆分
* 50万-100万数据以地区为单位拆分
* 10万-50万数据以县为单位查分
* 利用程序自动拆解数据，但需要明确每一个数据的字符编码
* 自动撰写元数据
* 自动撰写数据文档


#### 3.2.4 栅格数据的拆分和清洗原则
* 初步统一数据格式
 * geotif、hdf或netcdf或其他专业格式
 * arcinfo的数据格式建议转换为geotif格式
* 保持原有地图投影不变 
* 1km以上分辨率数据，建议以国家或省为单位切割
* 数据切割建议以矩形框为切割依据，不要用不规则多边形进行切割
* 全国数据切割通过自动方法进行
* 自动撰写元数据
* 自动形成缩略图
* 自动形成数据文档
