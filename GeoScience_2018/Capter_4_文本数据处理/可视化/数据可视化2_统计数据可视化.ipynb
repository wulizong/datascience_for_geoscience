### 2.统计数据随时间的变化

```python
df_new=df_GDP.set_index(['地区'])
df_new.head()
```

```python
year_list=list(range(2014,1948,-1))
year_list
```

```python
df_new.columns=year_list
df_new.head()
```

```python
df_t=df_new.T
```

```python
import matplotlib
font = matplotlib.font_manager.FontProperties(fname='C:\Windows\Fonts\simsun.ttc')

fig, ax = plt.subplots(figsize=[9, 6])
for region in df_new.index:
    p=df_t[region].plot(label=region)
    
ax.set_xlabel('Year')   
ax.set_ylabel('GDP')
ax.set_xlim([2014,1993])
ax.legend(df_new.index,loc='center right', bbox_to_anchor=(1.25,0.7),
          prop=font)
```

## 3.统计数据空间分布可视化

```python
shapefile='../Day_3/data/china_boundary_province.shp'
df_geo=gpd.read_file(shapefile,encoding='gbk')
df_geo.head()
```

```python
## 检查名字是否能对应上
```

```python
df_a=df_geo['NAME'].value_counts().reset_index()
df_a.columns
```

```python
df_b=df_GDP[u'地区'].value_counts().reset_index()
df_b.columns
```

```python
df_a.merge(df_b,how='outer',left_on=['index'],right_on=['index'])
```

```python
df_merged=df_geo.merge(df_GDP,how='inner',left_on='NAME',right_on=u'地区')
```

```python
df_2014=df_merged[['NAME','2014年','geometry']]
df_2014.head()
```

```python
%matplotlib inline
import matplotlib.pylab as plt
import pysal as ps
ax=df_2014.plot(figsize=(15,10),column=u'2014年',scheme='QUANTILES',k=9,linewidth=0.5)
#plt.colorbar(ax)
#plt.add_axes([0.4, 0.35, 0.015, 0.2])
```
