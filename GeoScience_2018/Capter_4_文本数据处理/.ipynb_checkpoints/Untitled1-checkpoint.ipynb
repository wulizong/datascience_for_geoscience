```python
import pandas as pd
%matplotlib inline
```

```python
filename_2003='data/DXG_JS2003.csv'
filename_2004='data/DXG_JS2004.csv'
filename_2005='data/DXG_JS2005.csv'
```

```python
col_names=['Id','Obs_Date','Precipitation']
```

```python
df_2003=pd.read_csv(filename_2003,names=col_names,header=0,index_col='Date',parse_dates={"Date":["Obs_Date"]})
df_2004=pd.read_csv(filename_2004,names=col_names,header=0,index_col='Date',parse_dates={"Date":["Obs_Date"]})
df_2005=pd.read_csv(filename_2005,names=col_names,header=0,index_col='Date',parse_dates={"Date":["Obs_Date"]})
```

```python
df_2003.columns,df_2004.columns,df_2005.columns
```

```python
ax=df_2003.plot(y='Precipitation')
ax1=df_2004.plot(ax=ax,y='Precipitation')
ax2=df_2005.plot(ax=ax,y='Precipitation')
```
